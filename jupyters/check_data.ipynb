```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
frame = pd.read_csv("/home/tanyx/dataDemo/campus2018/holiday_fix_diff_all.csv")
```

```python
frame[(frame["num_of_people"] == 0)&(frame["hour"]>5)&(frame["day"]==9)&(frame["hour"]==10)]
```

```python
month = 11
for loc_id in range(1,6):
    xs = []
    for hour in range(24):
        stds = 0
        for week_day in range(7):
            stds += frame[(frame["loc_id"] == loc_id)&(frame["month"] == month)&(frame["hour"] == hour)&(frame["week_day"] == week_day)]["num_of_people"].std()
        xs.append(stds)
    plt.subplot(1,5,loc_id)
    plt.plot(range(24), xs)
plt.title("month=%d"%(month))
plt.show()
```
